```python
# Generate random picture
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
resolution = (1448, 1923)

#random_vector = random.choice([0,1])
size = np.multiply(resolution[0], resolution[1])
```

```python
half_size = int(size / 2)
the_list = [0]*half_size + [1]*half_size
the_list
random_list = np.array(the_list)
#random_list = np.list(random_list)
```

```python
random.shuffle(random_list)
```

```python
random_list[:2]
```

```python
plt.imshow(random_list.reshape(resolution), cmap='Greys_r')
plt.xticks([]), plt.yticks([]) #
plt.title('Random Sample')
```
